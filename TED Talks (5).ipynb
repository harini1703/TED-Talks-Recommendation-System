# STEP-4 : Data Cleaning and Exploratory Data Analysis

1. The first line of code is converting the values in the 'film_date' column from Unix timestamps(Unix timestamps, also known as POSIX timestamps, are a way to represent dates and times as a single number) to a date format using the datetime module.


2. The apply() method is used to apply a lambda function to each value in the 'film_date' column. The lambda function converts the Unix timestamp (which is a number representing the number of seconds since January 1, 1970) to a datetime object using the fromtimestamp() method of the datetime module.fromtimestamp is a method in the datetime module in Python that creates a datetime object from a Unix timestamp. 


3. The strftime() method is then used to format the datetime object as a string in the format '%d-%m-%Y' (day-month-year).strftime() is a method in the datetime module in Python that formats a datetime object into a string. 


4. The second line of code is doing the same thing for the 'published_date' column.



```python
import datetime
tedmain['film_date'] = tedmain['film_date'].apply(lambda x: datetime.datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
tedmain['published_date'] =tedmain['published_date'].apply(lambda x: datetime.datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
```

1. The first line of code is creating a new column in the tedmain DataFrame called duration_hr, which represents the duration of each TED Talk in hours. The duration column contains the duration of each talk in seconds, so the duration value is divided by 60*60 (i.e., 3600) to convert it to hours.


2. The second line of code is converting the duration_hr column to a float data type using the astype() method (astype() is a method in pandas that is used to convert the data type of a pandas series or dataframe column to a specified type). This is necessary because the division in the previous step creates a column of floating-point numbers.


3. The third line of code is rounding the values in the duration_hr column to two decimal places using the round() method with the decimals parameter set to 2. 

```python
tedmain['duration_hr']=tedmain['duration']/(60*60)
tedmain['duration_hr']=tedmain['duration_hr'].astype(float)
tedmain['duration_hr']=tedmain['duration_hr'].round(decimals=2)
```

1. This code defines a Python function called get_top 


2. The function first selects a subset of columns from the tedmain DataFrame like the title of the talk, the name of the main speaker, the number of views, the number of comments, the date the talk was filmed, the date the talk was published, the duration of the talk in seconds, and the duration of the talk in hours. 


3. It then sorts the subset of the DataFrame in descending order based on the values in the column specified by the feature argument.by is a parameter,to specify the column(s) that should be used for sorting.


4. The sorted subset of the DataFrame is then indexed by the column specified by the feature argument and the resulting DataFrame contains the 10 talks with the highest values in the column specified by the feature argument.

```python
def get_top(feature):
    popular_talks = tedmain[['title', 'main_speaker', 'views','comments','film_date', 'published_date','duration','duration_hr']].sort_values(by=feature,ascending=False).set_index(feature).reset_index().head(10)
    return popular_talks
```

1. This is a Python function that creates a bar plot using the Seaborn library to show the top 10 TED Talks by a given feature. 

2. The function first creates a new DataFrame tedmain_top10 that contains the top 10 TED Talks by the specified feature. It then extracts the titles of these talks into a list called x_labels.

3. The function then creates a new figure with a size of 10x8 using plt.figure(), and creates a bar plot using sns.barplot() that shows the TED Talk titles on the x-axis and the specified feature on the y-axis.

4. The title, x-label, and y-label are also set using ax.set_title(), ax.set_xlabel(), and ax.set_ylabel(), respectively. The x-tick labels are rotated vertically using ax.set_xticklabels().

The function then uses a loop to add text labels to each bar in the plot, indicating the value of the specified feature for each TED Talk. This is done using ax.text(), which takes the x-coordinate, y-coordinate, text label, and horizontal and vertical alignment as arguments.

Finally, the function returns the bar plot.






```python
def get_graph(feature):
    tedmain_top10 = tedmain[['title', 'main_speaker', 'views','comments', 'published_date','duration','duration_hr']].sort_values(feature, ascending=False).head(10)
    x_labels=tedmain_top10['title'].to_list()
    plt.figure(figsize=(10,8))
    ax=sns.barplot(x='title',y=feature,data=tedmain_top10)
    ax.set_title('TOP 10 TED Talks by '+str(feature).upper())
    ax.set_xlabel('TITLE')
    ax.set_ylabel(feature)
    ax.set_xticklabels(x_labels,rotation='vertical')
    
    rects=ax.patches

    
    labels=tedmain_top10[feature].to_list()
    
    for rect,label in zip(rects,labels):
        height=rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2,height,label,ha='center',va='bottom')
    
```

* This is a Python code that uses the TfidfVectorizer class from the sklearn.feature_extraction.text module to convert the textual data in the description column of the tedmain DataFrame into numerical vectors.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3, strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',ngram_range=(1, 3),stop_words = 'english')
tedmain['description'] = tedmain['description'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(tedmain['description'])#Sparse
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
indices = pd.Series(tedmain.index, index=tedmain['title']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    idx = indices[title]

  
    sig_scores = list(enumerate(sig[idx]))

   
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

 
    sig_scores = sig_scores[1:11]


    title = [i[0] for i in sig_scores]

 
    return tedmain['description'].iloc[title]
```
